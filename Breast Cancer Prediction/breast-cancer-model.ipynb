# drop the Columns with high correlation to reduce multicollinearity



```python
# from the correlation matrix we got to know the worst columns are highly correlated with the mean columns.
# so we are droping these columns

# droping all worst columns

cols_worst = ['radius_worst', 
        'texture_worst', 
        'perimeter_worst', 
        'area_worst', 
        'smoothness_worst', 
        'compactness_worst', 
        'concavity_worst',
        'concave points_worst', 
        'symmetry_worst', 
        'fractal_dimension_worst']

df = df.drop(cols_worst, axis=1)
```

```python
# drop all columns to the perimeter and "area attributes"

cols_a = [
    'perimeter_mean',
        'perimeter_se', 
        'area_mean', 
        'area_se']
df = df.drop(cols_a, axis=1)
```

```python
# drp all the columns related to the concavity and conacave points" attributes
cols3 =['concavity_mean',
        'concavity_se', 
        'concave points_mean', 
        'concave points_se']
df = df.drop(cols3, axis=1)
```

## Train test split 

```python
X = df.drop(["diagnosis"], axis= 1)
y = df['diagnosis']
```

```python
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 0)
X_train.shape, X_test.shape
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
svc = svm.SVC()
svc.fit(X_train, y_train)
```

```python
# prediciton on the train and test data 
y_pred = svc.predict(X_test)
X_train_pred = svc.predict(X_train)
```

```python
# calculating the accuracy  of test data 
accuracy_score(y_test, y_pred)
```

```python
#calculating the accuracy of training  data
accuracy_score(y_train, X_train_pred)
```

```python
from sklearn.model_selection import GridSearchCV 
  
# defining parameter range 
param_grid = {'C': [0.1, 1, 10, 100, 1000],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['rbf']}
grid = GridSearchCV(svc, param_grid, refit = True, verbose = 3)

# fitting the model for grid search 
grid.fit(X_train, y_train) 
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
svc = svm.SVC(C=100, gamma= 0.01, kernel= 'rbf')
svc.fit(X_train, y_train)
```

```python
# prediciton on the training and test data  after hyperparameeter tuning
y_pred = svc.predict(X_test)
X_train_pred = svc.predict(X_train)
```

```python
# calculating the accuracy  of test data  after hyperparameeter tuning
accuracy_score(y_test, y_pred)
```

```python
#calculating the accuracy of training  data  after hyperparameeter tuning
accuracy_score(y_train, X_train_pred)
```
